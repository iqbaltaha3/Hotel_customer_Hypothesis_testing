#### Hypothesis Testing on Age:

Question: Is the average age of hotel guests significantly different from a specified value (e.g., 40 years)?

Test: One-sample t-test

### Rejection region approach

```python
# significance level = 0.05
# lets take a sample size of 30 
# specified value = 40
# h0 = The average age of hotel guests is equal to the specified value (e.g., 40 years).
# h1 = The average age of hotel guests is significantly different from the specified value.
```

```python
sample = np.array(hotel['Age'].sample(30,random_state=42)) # taking a sample of 30 values 
sample_mean = sample.mean()  # finding sample mean
sample_mean
```

```python
sample_std = np.std(sample,ddof=1) # finding standard deviation
sample_std
```

```python
specified_value = 40
```

```python
# Calculating t-statistics
tstat = (sample_mean - specified_value)/(sample_std/(len(sample)**0.5))
tstat
```

```python
# now we look at the critical value for corresponding tstat in t-table
dof = len(sample)
alpha = 0.05

from scipy.stats import t

critical_value = t.ppf((1-alpha)/2,dof)
critical_value
```

```python
if np.abs(tstat) > critical_value:
    print("Reject Null Hypothesis: The average age of hotel guests is significantly different from", specified_value, "years.")
else:
    print("Fail to Reject Null Hypothesis: The average age of hotel guests is not significantly different from", specified_value, "years.")
```

### P-value Approach

```python
# Calculate p-value
p_value = 2 * (1 - t.cdf(abs(tstat), dof))

if p_value < alpha:
    print("Reject Null Hypothesis: The average age of hotel guests is significantly different from", specified_value, "years.")
else:
    print("Fail to Reject Null Hypothesis: The average age of hotel guests is not significantly different from", specified_value, "years.")
```

If the p-value approach and critical region approach support opposite hypotheses in the same problem, it indicates a discrepancy or inconsistency between the two methods. This situation is rare but could occur due to several reasons, such as:

Different Assumptions: The p-value approach and critical region approach are based on different assumptions or statistical properties of the test statistic. If these assumptions are violated or if the sample size is small, the results from the two approaches may diverge.

Type of Test: The p-value approach focuses on the probability of observing the test statistic or more extreme values under the null hypothesis, while the critical region approach defines rejection regions based on critical values from a probability distribution. If the test is one-tailed in one approach and two-tailed in the other, the conclusions may differ.

Calculation Errors: Errors in calculating the test statistic, critical values, or p-values can lead to discrepancies between the two approaches.

Sampling Variability: Due to random sampling variability, there can be differences in the results obtained using different statistical methods. However, these differences should be within an acceptable range of variation.

If the p-value approach and critical region approach support opposite hypotheses, it's essential to carefully examine the assumptions, calculations, and interpretation of results to identify the source of discrepancy. In such cases, it may be necessary to reconsider the analysis, verify calculations, or explore alternative statistical methods to ensure the reliability of the conclusions drawn from the hypothesis test. Consulting with a statistician or peer review can also help in resolving such discrepancies.

### Hypothesis Testing on Lodging Revenue:

Question: Is there a significant difference in lodging revenue between guests from different nationalities (e.g., comparing revenue between guests from Germany and France)?

Test: Independent samples t-test (assuming equal variances) or Welch's t-test (if variances are unequal)

```python
# h0 = No difference
# h1 = Significant difference
```

```python
fra_LodgingRevenues = hotel[hotel['Nationality'] == 'FRA']['LodgingRevenue']
ger_LodgingRevenues = hotel[hotel['Nationality'] == 'DEU']['LodgingRevenue']
```

```python
fra_sample = np.array(fra_LodgingRevenues.sample(n=30,random_state=42))
ger_sample = np.array(ger_LodgingRevenues.sample(n=30,random_state=42))
```

#### Checking for similarity in variance (levene)

```python
from scipy.stats import levene

# Perform Levene's test for equality of variances
statistic, p_value_var = levene(ger_sample, fra_sample)

# Set significance level
alpha_var = 0.05

# Make decision based on p-value
if p_value_var < alpha_var:
    print("Reject Null Hypothesis: Variances are significantly different.")
else:
    print("Fail to Reject Null Hypothesis: Variances are not significantly different.")
```

#### Checking for Normality (Shapiro-Wilk test)

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for normality
statistic_germany, p_value_germany = shapiro(ger_sample)
statistic_france, p_value_france = shapiro(fra_sample)

# Make decision for Germany
if p_value_germany < alpha:
    print("Reject Null Hypothesis for Germany: The data does not follow a normal distribution.")
else:
    print("Fail to Reject Null Hypothesis for Germany: The data follows a normal distribution.")

# Make decision for France
if p_value_france < alpha:
    print("Reject Null Hypothesis for France: The data does not follow a normal distribution.")
else:
    print("Fail to Reject Null Hypothesis for France: The data follows a normal distribution.")
```

#### Performing 2-sample t_test (using ttest_ind)

```python
# Rejection Region Approach
from scipy.stats import ttest_ind 

t_statistic, p_value = ttest_ind(ger_sample, fra_sample, equal_var=False)  # Assuming unequal variances
alpha = 0.05
critical_value = t.ppf(1 - alpha/2, min(len(ger_sample) - 1, len(fra_sample) - 1))

if abs(t_statistic) > critical_value:
    print("Reject Null Hypothesis: There is a significant difference in lodging revenue between guests from Germany and France.")
else:
    print("Fail to Reject Null Hypothesis: There is no significant difference in lodging revenue between guests from Germany and France.")
```

```python
# P-value Approach
if p_value < alpha:
    print("Reject Null Hypothesis: There is a significant difference in lodging revenue between guests from Germany and France.")
else:
    print("Fail to Reject Null Hypothesis: There is no significant difference in lodging revenue between guests from Germany and France.")
```
